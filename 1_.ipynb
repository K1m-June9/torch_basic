수치미분 구현(파이토치의 미분 방법)

```python
def fdiff(f):
    #함수 f를 인수로 미분한 결과 함수를 diff로 정의
    def diff(x):
        h = 1e-6
        return (f(x+h) - f(x-h)) / (2*h)
    
    #fdiff의 반환은 미분환 결과 함수 diff
    return diff
```

```python
# 2차 함수의 수치 미분

#f의 미분 결과 함수 diff를 취득
diff = fdiff(f)

#미분 결과를 계산하고 y_dash에 대입
y_dash = diff(x)

print(y_dash)
```

```python
#결과 그래프 출력

plt.plot(x, y, label = r'y = f(x)', c = 'b')
plt.plot(x, y_dash, label = r"y = f'(x)", c = 'k')
plt.legend()
plt.grid()
plt.show()
```

시그모이드 함수

### $g(x)$ = $1 \over{1 + exp(-x)}$

```python
#시그모이드 함수 정의
def g(x):
    return 1 / (1 + np.exp(-x))
```

```python
y = g(x)
print(y)
```

```python
#시그모이드 그래프
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
# 시그모이드 함수의 수치 미분

diff = fdiff(g)

y_dash = diff(x)

print(y_dash)
```

```python
#결과그래프

plt.plot(x, y, label = r'y = g(x)', c = 'b')
plt.plot(x, y_dash, label = r"y = g'(x)", c = 'k')
plt.legend()
plt.grid()
plt.show()
```
